```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
COMP = pd.read_csv("MNIST_TIME(Comp).csv")
JET = pd.read_csv("MNIST_TIME(jetson).csv")
SEV = pd.read_csv("MNIST_TIME(server).csv")
```

```python
# 차트 그리기
fig = plt.figure(figsize=(20, 12)) # 차트 생성 및 사이즈 설정
ax = fig.add_subplot(1,1,1) # subplot 생성

ax.plot(COMP['0'], label='Computer', color='r') # CLOU 종가 불러오기
ax.plot(JET['0'], label='Jetson Nano', color='b') # SKYY 종가 불러오기
ax.plot(SEV['0'], label='Server', color='g') # SKYY 종가 불러오기

ax.set_title('Computer & Jetson Nano Time & Server', fontsize=20) # 타이틀 설정
ax.set_ylabel('Time', fontsize=14) # x축 설정
ax.set_xlabel('batch_index', fontsize=14) # y축 설정

ax.legend(fontsize=12, loc='best') # 범례 설정 best로 해놓으면 가장 적절한 위치에 알아서 범례가 놓이게 됩니디

plt.show()
```
